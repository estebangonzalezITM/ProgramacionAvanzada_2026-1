## Tipos de datos básicos

Los principales tipos de datos básicos son:

1. Representacion Hex, Binario, Decimal
2. Numéricos
    - Entero (int)
    - Punto flotante (float) https://es.wikipedia.org/wiki/IEEE_754
    - Números complejos (complex)
2. Textos
    - Cadena de caracteres (str)
3. Booleanos
    - True o False

```python
a = 0xFF            # Valor hexadecimal FF (255 en decimal)
print(a)

a1 = int("FF", 16)  # Valor hexadecimal FF (255 en decimal)
print(a1)

b = hex(240)        # Valor decimal 255 (FF hexadecimal)
print(b)
```

```python
c = 0b11111111            # Valor binario 11111111 (255 en decimal)
print('c =',c)

c1 = int("11110111", 2)   # Valor binario 11111111 (255 en decimal)
print('c1 =',c1)

d = bin(240).zfill(8)     # Valor decimal 255 (11111111 en binario)
print('d =',d)
```

```python
e = 20653     # Valor entero
f = 2.45      # Valor flotante
g = 2 + 4j    # Valor complejo
```

```python
h = 'a'             # String
k = 'Hola mundo'    # String
k1 = "Hola mundo!"  # String

print(k1)
```

```python
l = True
print(l)

l1 = False
print(l1)
```

La función __type__ de Python, nos permite preguntar por el tipo de dato de una variable.

## Todo en Python son objetos

Todo en Python es un objeto, esto es, todo dato tiene:

1. Atributos.
2. Funciones (métodos).

Podemos consultar los métodos de un tipo de datos, accediendo a la documentación de Python. Por ejemplo, para los Strings: https://docs.python.org/3.7/library/string.html

```python
# Los atributos son las características o el estado del objeto, por ejemplo un número complejo tiene parte real e imaginaria

numComplejo = 4 + 3j

print(numComplejo.real)          # .real es un atributo del tipo de datos Complex. Atributo: 4
print(numComplejo.imag)          # .imag es un atributo del tipo de datos Complex. Atributo: 3
print(numComplejo.conjugate())   # .conjugate es un método del tipo de datos Complex.
```

```python
msg = "Lorem ipsum dolor sit amet, consectetur adipiscing elit. Sed in tortor vitae nisi fermentum porta"
print(msg)
print(msg.replace('i','t'))  # .replace es un metodo del tipo de datos String
print(msg)                   # Cuidado, crea un nuevo objeto, no sobreescribe el objeto original

msg = msg.replace('i','t')   # Ahora se reescribe el objeto msg
print(msg)
```

### Paso por referencia (punteros)

Por defecto cuando se realiza una asignacion entre variables en Python, no se crea un espacio nuevo en memoria, sino que la nueva variable `apunta' al mismo espacio de la variable originial.

```python
a = 346
b = a

print("comparacion si estan en el mismo espacio de memoria: ",a is b)
print("Ids: a:",id(a)," b:",id(b))
```

## Palabras Reservadas

Todo lenguaje de programación tiene palabras que son de uso privativo del mismo, es decir, que **no es conveniente** y en algunos casos es **imposible** utilizarlas, por ejemplo:

- sqrt, abs, etc.
- sin, cos, tan, etc.
- for, while, if, else, etc.
- True, False


Algunas palabras no pueden/deben ser utilizadas en la definición de elementos dentro del programa, pues Python las conserva reservadas para ciertas funciones. Algunos casos son: *if, for, while, import, from, True, False, etc*: https://www.programiz.com/python-programming/keywords-identifier

```python
# abs = 3
a = abs(4 + 3j)
print(a)
```

## Precedencia de operadores

__Recordemos__ que la jerarquía de operadores aritméticos es la siguiente:

1. $()$: parentesis.
2. $**$ : potencia.
3. $* /$ % : multiplicación, división y módulo.
4. $+ -$ : suma y resta.

```python
help(print)
```

```python
a = -5
b = 10

c = a + b + 4 #suma de dos variables a y b mas un literal 4
# print('c es', c, "esta es la suma de a + b:", a, '+', b, sep=' ')
```

```python
d = a / b #division
print('d es',d)
```

```python
e = 2**4 # Potencia 2^4
print('e es',e)
```

```python
a = 2**(2+4)
print('a es',a)
```

```python
g = 17 / 2 # División
print('g es',g)
```

```python
g = 17 // 2   # División de piso (dividir los números y redondear el resultado hacia abajo)
print('g es',g)
g2 = -17//2
print('g2 es',g2)
```

```python
f = 17 % 2       # Módulo o residuo (el resultado es el sobrante o resto de esa división)
print('f es',f)
```
