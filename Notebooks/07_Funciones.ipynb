## **1.1 Declaración de funciones**


En Python se utiliza la palabra reservada ` def ` para definir una función. La sintaxis es la siguiente:

<br>

 **def** _nombreDeLaFuncion_ **(** $parametro_1, parametro_2, ..., parametro_n$ **):**

A continuación un ejemplo:

```python
def imprimirMensaje(nombre="Estudiante"):
    '''
    Docstring.
    Procedimiento que imprime un mensaje utilizando un valor de entrada.

    INPUT:
        nombre: str
    OUTPUT:
        None
    '''
    if(type(nombre) != str): #validamos que el argumento <nombre> sea de tipo string
        print("Error, el parámetro de entrada debe ser un str.")
    else:
        print(f"Hola {nombre}. Bienvenido al curso")
```

```python
print(imprimirMensaje())
```

```python
 # El parámetro de la función es *nombre*. *Estudiante* sería el valor por defecto del parámetro,
 # El valor por defecto es opcional (pero recomendado), y se usa cuando no se le entrega un valor
 # a la función, entoces utilizará el valor por defecto.
```

```python
help(imprimirMensaje)
# imprimirMensaje?
```

```python
#llamado a la funcion
s = 'Miguel'

imprimirMensaje(s)
```

```python
imprimirMensaje('Esteban')
```

```python
imprimirMensaje(98.86)
```

## **1.2 Parámetros y argumentos**

Definiciones:

1. **Parámetros**: Variables de la definición de la función y a las que son asignadas un valor al momento de llamar la función.
2. **Argumentos**: Valores que son asignados a los parámetros de la función al momento del llamado.

El número de parámetros que puede recibir una función es arbitrario, es decir, tantos como se necesiten.

```python

def miFuncion(a,b,c,d): # Parametros de la funcion: a, b, c, d (notar que no tienen valores por defecto)
    '''
    Función que recibe cuatro parámetros y retorna el cuadrado del producto de ellos
    '''
    f = (a*b*c*d)

    return f**2  # *return* es la instrucción que permite que la función "entregue" un resultado al programa que la llamó.

miFuncion(1,2,1,1)
```

```python
resultado = miFuncion(1,2,3,1) # argumentos de la funcion
print(resultado)
```

* Los parámetros y argumentos son **Variables Locales**:

En el ejemplo, las variables `a`, `b`, `c`, `d` y `f` son locales a la función. Nacen cuando la función empieza a ejecutarse y mueren (se borran de la memoria) en cuanto la función llega al return o termina.

## **1.3 Parámetros por defecto**

Python permite la asignación de valores por defecto a los parámetros de una función. Esto es, es posible asignar un valor por defecto a uno o varios parámetros de la función, de tal forma que si el usuario no ingresa un argumento para un parámetro, se asigna el valor por defecto a este último.

```python
# ¿Qué pasa si no tenemos valores por defecto?
def miFuncion(a,b,c,d):
    f = (a*b*c*d)
    return f**2

miFuncion(1,2,1)
```

```python
def miFuncion(a,b=1,c=1,d=2): #parametros de la funcion: a, b, c, d
    '''
    Función que recibe cuatro parámetros y retorna el cuadrado producto de ellos
    '''
    f = (a*b*c*d)
    return f**2

resultado = miFuncion(2) #argumentos de la funcion
print(resultado)
```

```python
def y(v0, t, g=9.8): #valor por defecto para el parámetro g
    y = v0*t - 0.5*g*t**2
    return y
y(10,1,9.78)
```

```python
for t in range(1,5):
    print("tiempo: "+str(t)+" s ; y = "+str(y(5,t)))
```

```python
for t1 in range(1,5):
    print("tiempo: "+str(t1), y(t = t1, v0 = 5)) #llamada a la funcion utilizando el valor por defecto para g
```

## **1.4 Uso de**  `*args` **y** `**kwargs`

¿Para qué se usan? En realidad no tienes porque usar los nombres `args` o `kwargs`, ya que se trata de una mera convención entre programadores.

Sin embargo lo que si que tienes que usar es el asterisco simple `*` o doble `**`. Es decir, podrías escribir `*datos` y `**numeros` y el resultado será el mismo.

https://python-intermedio.readthedocs.io/es/latest/args_and_kwargs.html

```python
def test_var_args(*argv):
    for arg in argv:
        print("argumentos de *argv:", arg)
    print(argv)
    print(type(argv))
```

```python
#test_var_args('python', 'foo', 'bar')
test_var_args(1,10,100)
```

* Como acabamos de ver, los datos guardados con `*args` se guardan como una tupla

* Puede usarse para guardar datos cuando no se sabe cuantos se van a recibir, o cuando hay datos opcionales:

```python
def test_var_args(f_arg, *argv):
    print("primer argumento normal:", f_arg)
    print(f_arg)
    print(type(f_arg))
    print(argv)
    print(type(argv))

#test_var_args('python', 'foo', 'bar')
# test_var_args(1,10,100)
```

```python
def test_var_args2(*datos):
    suma = sum(datos)
    print (f'Los datos ingresados son {datos}')
    return 'Los datos ingresados suman '+ str(suma)

test_var_args2(1,20,100,1000)
```

La diferencia fundamental entre `*args` y `**kwargs` radica en cómo empaquetan los datos que sobran al llamar a una función.

Si `*args` es una "bolsa" de valores, `**kwargs` es un "contenedor etiquetado", que permite pasar argumentos con un nombre asociado (llave=valor).

`**args` guarda los datos en una tupla y `**kwargs` en un diccionario

```python
def saludame(**kwargs):
    #for key, value in kwargs.items():
    #    print(f"{key} = {value}")
    print(kwargs)
    print(type(kwargs))

saludame(nombre="Covadonga", apellido = "Verdugo")
```

```python
def saludame(**kwargs):
    print(kwargs)
    print(type(kwargs))
```

## ¿Para qué sirven en la vida real?

`*args` es útil para funciones matemáticas que reciben muchos números o para imprimir mensajes variables.

`**kwargs` es el estándar en librerías de visualización (como Matplotlib) o de Machine Learning (como Scikit-Learn). Se usa para pasar configuraciones y parámetros (como el color, el tamaño de la fuente, el número de neuronas, etc.) de forma opcional.

## **1.5 Valores de retorno**

Una función puede retornar un valor. En cuyo caso utilizamos la palabra reservada __return__.

Sintaxis:
    
__return__ objeto_a_retornar


```python
def cuad(x):
    return x**2 #se retorna el valor x al cuadrado

x = 15

Respuesta = cuad(x)

print(Respuesta)
print(type(Respuesta))
print(cuad(x))
```

* Si solo se imprime (`print`), la función muestra el resultado, pero tú se puede usar ese número para otra operación matemática fuera de ella.

* Si se usa `return`, la función "entrega" resultado para que puedas guardarlo en una variable y seguir trabajando con él.

```python
def cuad2(x):
  print(x**2)
Rta = cuad2(x)
# print(type(Rta))
# print(Rta)
```

* Una función con valor de retorno, equivale en el momento del llamado, al tipo de dato que retorna. Es decir, una función que devuelve o retorna un valor, soporta todas las operaciones del tipo de dato que retorna.

```python
import math

x = 1
y = math.sqrt(cuad(x) + 3.24) #estamos utilizando cuad(x) como un flotante

print(y)
```

## **1.6 Múltiples valores de retorno**

En caso de que una funcion deba retornar más de un valor, podemos:

1. __Agregar los elementos__ a retornar en una estructura de datos.
2. __Retornar los elementos separados por coma (,)__: En este caso, Python se encarga de agregar automáticamente los elementos en una tupla.

```python
## definir una función que reciba dos números como parámetros, debemos
## retornar la suma, la resta, el producto y la división de los
## dos números

def ejemplo(*args):
    num1, num2, *_ = args
    suma = num1+num2
    resta = num1-num2
    producto = num1*num2
    division = num1/num2
    print(f'Los números ingresados son {args}')
    lista = [suma, resta, producto, division]

    return lista
```

```python
ejemplo(1,2)
```

```python
a = 3
b = 2

L = ejemplo(a,b)
print(type(L))
print(L)
print(L[2])
```

```python
## definir una función que reciba dos números como parámetros, debemos
## retornar la suma, la resta, el producto y la división de los
## dos números

def ejemplokwargs(**args):
    num1 = args['e1']
    num2 = args['e2']
    suma = num1+num2
    resta = num1-num2
    producto = num1*num2
    division = num1/num2
    print(f'El diccionario ingresado es {args}')
    lista = [suma, resta, producto, division]

    return lista
```

```python
ejemplokwargs(e1=1,e2=2)
```

```python
def ejemplo2(num1,num2):
    suma = num1+num2
    resta = num1-num2
    producto = num1*num2
    division = num1/num2

    dic = {"s":suma, "r": resta, "p":producto, "div":division}

    return dic
```

```python
dic = ejemplo2(1,2)
print(dic['s'])
```

```python
a = 3

L = ejemplo2(a,2)
print(type(L))
print(L)
print(L["s"])
```

## **Localidad y globalidad de una variable**

Una variable puede ser local o global respecto a una función, i.e., el valor de la variable puede existir de forma interna o externa a la función. Por definición, todo parámetro de una función es local a la misma.

```python
c = 10 #c es una variable por fuera de la funcion suma, esto es, c es global

def suma(a, b):
    c = a + b #la variable c es local respecto a la funcion suma
    return c

print("Valor de c dentro de la funcion: ", suma(2, 4))
print("Valor de c por fuera de la funcion: ", c)
```
