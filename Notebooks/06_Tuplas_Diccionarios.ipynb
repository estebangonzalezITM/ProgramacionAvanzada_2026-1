# **1. Tuplas**

## **1.1 Definición**

Las tuplas son una estructura de datos integrada en Python que permite almacenar una colección de elementos de forma ordenada. Se parecen mucho a las listas, pero tienen una diferencia fundamental que las define: **son inmutables**.

Para crear una tupla, los elementos se encierran entre <font color="blue" size=3>paréntesis () <font color="black" size=3>y se separan por comas. Aunque en muchos casos los paréntesis son opcionales, se recomienda usarlos por claridad.

* Verificar tipo

```python
tupla1 = (10, 20, 5, 30)
print((type(tupla1)))
```

* Ver un elemento de una tupla

```python
tupla3 = (1, "Hola", 3.14, True)
print(tupla3[2])
```

* Las entradas de una tupla pueden ser mixtas

```python
tupla4 = ( 20 , 'string' , 3.5 , 85+4j , [30,60,10],  50 , 'texto',20 )
tupla4
```

## **1.2 Funciones escenciales con tuplas**

`len(tupla)`: Devuelve el tamaño o cantidad de elementos. Fundamental para ciclos for.

`max(tupla)` y `min(tupla)`: Devuelven el valor más alto y más bajo (siempre que los datos sean comparables, como números).

`sum(tupla)`: Suma todos los elementos numéricos.

`sorted(tupla)`: ¡CUIDADO!: Esta función ordena los elementos de la tupla, pero como la tupla es inmutable, devuelve una lista nueva con los elementos ordenados.

```python
tupla1 = (10, 20, 5, 30)
tupla2 = ("rojo", "verde", "azul")
tupla4 = ( 20 , 'string' , 3.5 , 85+4j , [30,60,10],  50 , 'texto',20 )
```

```python
print(len(tupla1))
```

## **1.3 Métodos con tuplas**

Las tuplas tienen solo tienen dos métodos propios. Esto se debe a su naturaleza inmutable: como no se pueden cambiar, agregar o eliminar elementos una vez creadas, no necesitan métodos de modificación.

**Método `.count(valor)`**

Este método cuenta cuántas veces aparece un elemento específico dentro de la tupla. Es útil para validaciones estadísticas o de frecuencia.

```python
mi_tupla = (1, 2, 3, 1, 1, 4)
print(mi_tupla.count(1))
```

**Método `.index(valor)`**

Busca un elemento en la tupla y devuelve el índice (la posición) de su primera aparición. Si el elemento no se encuentra, Python lanza un error (*ValueError*).

```python
colores = ("rojo", "verde", "azul", "verde")
print(colores.index("verde"))
```

* Recorrer una tupla

```python
tupla3 = ( 1.3 , 2.2 , 7 , 4.1 , 4.5E-005 , "hola" , "mundo" , ["a", "b", "c"] )

print("Longitud de la tupla: ",len(tupla3),'\n')

for i in range(len(tupla3)):
    print("Posición",i,': ', tupla3[i])
```

# **2. Diccionarios**

## **2.1 Definición**

En Python, los diccionarios son una estructura de datos que permite almacenar información en pares <font color="blue" size=3> llave → valor.

Son útiles cuando quieres acceder a un dato usando un identificador en lugar de una posición numérica (como ocurre en las listas), con la siguiente sintaxis:

$diccionario$ = <font color="blue" size=5>{</font> $key_1:elemento_1$, $key_2:elemento_2$, $key_3:elemento_3$,...,$key_n:elemento_n$ <font color="blue" size=5>}</font>

Donde $key_i$ se refiere a un indicador, llave o clave única para el elemento $i$.

* Observemos el siguiente ejemplo:

```python
persona = { "nombre": "Esteban" , "edad": 35 , "profesion": "Docente" }
print(persona)
print(type(persona))
```

<font color="red">"nombre" <font color="black">es la clave, <font color="red">"Esteban" <font color="black">es el valor

<font color="red">"edad" <font color="black">es la clave, <font color="red">"35" <font color="black">es el valor

<font color="red">"profesion" <font color="black">es la clave, <font color="red">"Docente" <font color="black">es el valor

*   Para acceder a un valor se usa la llave (key):

```python
print(persona["profesion"])
```

Un diccionario en Python es como un diccionario real, cuando buscas una clave (palabra), obtienes su valor (definición)

```python
# Una lista equivalente sería
# ¡Cuidado! A diferencia de los diccionarios, para llamar cada elemento debes conocer su posición en la lista
persona_list = ["Esteban",35,"Docente"]
print(persona_list[0])
```

* Es posible crear un diccionario vacío:

```python
mi_diccionario = {}
print(mi_diccionario)
```

* Los diccionarios también pueden crearse con el constructor `dict` (aunque la sintaxis puede ser un poco más compleja)

```python
mi_diccionario = dict()
print(mi_diccionario)
```

```python
# Es equivalente a hacer:
# persona = { "nombre": "Esteban" , "edad": 35 , "profesion": "Docente" }

persona = dict( [("nombre", "Esteban") , ("edad", 30) , ("profesion", "Docente")])
persona
```

<font color="red"> ¡¡¡ CUIDADO !!!

Si se usan llaves pero sin pares **clave → valor**, Python crea un set:

```python
num_set = {1, 2, 3, 4, 1, 2}
print('Este es un set: ',num_set)
print(type(num_set))
```

Un *set* **NO es un diccionario**, es un conjunto (no permite elementos repetidos).

* Las llaves pueden ser de tipo *string, int, float, complex, tuple*
* Las llaves NO pueden de tipo *list, dict, set* (mutables)

```python
num_dic = { 'uno':1 , 2:'dos' , 3.0:'tres' , 4+4j:'cuatro' , (5,5):5 }
print('Este es un diccionario: ',num_dic)
```

* Los atributos y métodos pueden consultarse con el comando:

```python
dir(dict)
```

## **2.3 Métodos de los diccionarios**

* Asumir que tenemos un diccionario definido como  **dic = {'x': 10, 'y':20}**.

| Metodo  | Descripcion  | Salida |
|---|---|---|
| .keys()  | Retorna las llaves de un diccionario.  | dict_keys(['x', 'y'])|
| .values()  | Retorna los valores asociadas a cada llave del diccionario.  | dict_values([10, 20])|
| .items()  | Retorna en tuplas (llave$_i$,value$_i$)  | dict_items([('x', 10), ('y', 20)]) |
| .get('x')  | Retorna el valor asociado a la llave.   | 10 |
| .get('z')  | Si la llave no existe devuelve un *None* por defecto  | None |
| .get('w',0)  | O devuelve el segundo argumento su la llave no existe.  | 0 |
| .pop('x')  | Retorna el valor asociado a la llave, <br> y la llave sale del diccionario.  | 10; dic = {'y':20} |
| .popitem()  | Retorna la ultima llave con su valor en una tupla, <br> y la ultima llave sale del diccionario.  | 20; dic = {'x':10} |
| .update({'x':15}) | Busca la llave y modifica el valor en el diccionario.  | None; dic = {'x':15, 'y':20} |
| .update({'z':30}) | Si la llave no existe, agrega la llave y el valor al diccionario.  | None; dic = {'x':10,'y':20,'z':30} |
| .setdefault('y',50) | Retorna el valor de la llave si la llave esta en el diccionario, <br> de lo contrario devuelve el valor por defecto.  | 20; dic = {'x':10,'y':20} |
| .setdefault('z',50) | Retorna el valor de la llave si la llave esta en el diccionario, <br> de lo contrario devuelve el valor por defecto.  | 50; dic = {'x':10,'y':20,'z':50} |
| .clear() | Limpiar el diccionario.  | {} |
| .copy() | Retornar una copia del diccionario.  | {'x':10, 'y':20} |

```python
dic = {'x': 10, 'y':20}
metodo = dic.popitem()

print(' Metodo: ',metodo)
print('\n Diccionario: ',dic)
```

* Dependiendo del método, la salida puede convertirse en una lista, usando `list()`

```python
dic = {'x': 10, 'y':20}

print(' Metodo: ', dic.values() )

valores_lista = list( dic.values() )
print('\n Lista: ',valores_lista)
```

* ¿Como saber si una llave existe?

```python
Dic = {"primero":1,"segundo": 2, "letra": 'a', 5.1 : [50,5,2,1,6]}
a = 5.1
if a in Dic.keys():
    print(f"La llave {a} si existe!")
else:
    print(f"La llave {a} no existe!")
```

* Creando llaves nuevas en diccionarios `dic['llave nueva']` = `valor_nuevo`

```python
Dic = {"primero":1,"segundo": 2, "letra": 'a', 5.1 : [50,5,2,1,6]}
Dic['llave nueva'] = [1,2,3,4,5]
print(Dic)
Dic[10] = [1,2,3,4,5]
print(Dic)
```

* Llenando un diccionario

Para ello analicemos el siguiente ejemplo:

```python
import math

tabla_factorial = {} #diccionario vacio
print(tabla_factorial)

for i in range(15):
    tabla_factorial[i] = math.factorial(i)

print(tabla_factorial)
```

```python
print(tabla_factorial[8])
```

```python
for key, valor in tabla_factorial.items():
    print(f"factorial de {key} es {valor}")
```
