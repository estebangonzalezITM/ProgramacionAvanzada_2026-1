
# 2. Operadores Relacionales en Python


- Se usan para establecer relación entre valores (variables)
- Producen un resultado booleano (True o False):

Una expresión Booleana se puede definir como toda expresión que retorna un valor falso o verdadero:
     
$ expresion \rightarrow [True, False] $

- Todos tienen el mismo nivel de precedencia (prioridad), pero son de menor prioridad que los operadores aritméticos

| Operador | Operación |
| --- | --- |
| > | Mayor que |
| < | Menor que |
| >= | Mayor o igual que |
| <= | Menor o igual que |
| != | Diferente  |
| == | Igual |

```python
# Celda para ejemplos de operadores relacionales

a = 10
b = 3

variable = a >= b

print(variable)
print(type(variable))
```

```python
# Los operadores se pueden encadenar
print(0 < a < 10)
```

```python
x = 0 < a < 10
print(x)
```

# 2. Operadores Lógicos en Python


- Se usan para establecer relación entre valores (variables)
- Producen un resultado booleano (True o False):

| Operador | Operación | Resultado |
| --- | --- |--- |
| and | Y (conjunción) | True cuando ambas condicones son verdaderas |
| or | O (disyunción) | True cuando al menos una condición es verdadera |
| not | No (negación) | Invierte el valor |


```python
a = 3

print( a > 10 and a < 20)

print( a > 1 and a < 20)
```

```python
print( a > 10 or a < -10)

print( a > 10 or a < 10 )
```

```python
print(not( a > 10 or a < -10))

not( a > 10 or a < 10)
```

# 3. Estructura condicionales

En Python contamos con tres elementos condicionales de la siguiente forma:

1. **if**(_expresion booleana_)**:** --> Si se cumple la expresión booleana.
2. **else:** --> en caso contrario.
3. **elif**(_expresion booleana_)**:** --> Si no se cumple la expresión anterior y se cumple esta. Viene a reemplazar el `if` anidado de un nivel.


# IF

Si la condición que pones al lado es verdadera (True), el código dentro de él se ejecuta. Si no, se lo salta.

<br>

<font color="#130eb3" size=5>__if__</font> <font size=5>(condición)</font> <font color="#130eb3" size=5>__:__</font>

    instrucción 1
    
    instrucción 2
    
    ...
    
    instrucción n


```python
#Ejemplo if

edad = float(input("Ingrese su edad en años: "))

if(edad >= 18):
    print("Eres mayor de edad")
    print("instruccion 2")

print("Esto es fuera del if")
```


# IF - ELSE

Si no se cumple la condición del IF (o cualquier otra condición anterior), se ejecuta automáticamente. Es opcional, pero útil para cerrar el ciclo.

<br>

<font color="#130eb3" size=5>__if__</font> <font size=5>(condición)</font> <font color="#130eb3" size=5>__:__</font>

    instrucción 1
    
    instrucción 2
    
    ...
    
    instrucción n
    
<font color="#130eb3" size=5>__else:__</font>

    instrucción sino 1
    
    instrucción sino 2
    
    ...
    
    instrucción sino n

No lleva ninguna condición porque se ejecuta automáticamente si ninguna de las condiciones anteriores fue verdadera. Es opcional, pero útil para cerrar el ciclo.

```python

#Ejemplo if - else

edad = float(input("Ingrese su edad en años: "))

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("No eres mayor de edad")

print("Esto es fuera del if-else")
```

# ELIF (ELSE - IF )

Se usa cuando quieres revisar una segunda, tercera, cuarta... condición(es), pero solo si las anteriores fallaron. Puedes tener tantos elif como necesites.

<br>

<font color="#130eb3" size=5>__if__</font> <font size=5>(condición)</font> <font color="#130eb3" size=5>__:__</font>

    instrucción 1
    
    instrucción 2
    
    ...
    
    instrucción n
    
<font color="#130eb3" size=5>__elif__</font> <font size=5>(condición2)</font> <font color="#130eb3" size=5>__:__</font>

    instrucción condición2 1
    
    instrucción condición2 2
    
    ...
    
    instrucción condición2 n
    
<font color="#130eb3" size=5>__else:__</font>

    instrucción sino 1
    
    instrucción sino 2
    
    ...
    
    instrucción sino n

```python
#Ejemplo elif - else

edad = float(input("Ingrese su edad en años: "))

if (edad < 18):
    print("No eres mayor de edad")
elif (edad >= 60):
    print("Eres un adulto mayor")
else:
    print("Eres mayor de edad")

print("Esto es fuera del if-else")
```

```python
# Ejemplo if-elif-else
# Solicitar coordenadas de un punto y decir en qué cuadrante del plano cartesiano se encuentra
# (x, y)

x = float(input("ingrese el valor de x: "))
y = float(input("ingrese el valor de y: "))

if (x>0):
    if (y<0):
        print("El punto está en el cuadrante 4")
    else:
        print("El punto está en el cuadrante 1")
else:
    if (y<0):
        print("El punto está en el cuadrante 3")
    else:
        print("El punto está en el cuadrante 2")
```

```python
# Ejemplo if-elif-else
# Solicitar coordenadas de un punto y decir en qué cuadrante del plano cartesiano se encuentra
# (x, y)

x = float(input("ingrese el valor de x: "))
y = float(input("ingrese el valor de y: "))

if (x > 0 and y > 0):
    print("El punto está en el cuadrante 1")
elif (x < 0 and y > 0):
    print("El punto está en el cuadrante 2")
elif (x < 0 and y < 0):
    print("El punto está en el cuadrante 3")
elif (x == 0 and y == 0):
    print("El punto está en el origen")
elif (x == 0):
    print("El punto está sobre el eje horizontal")
elif (y == 0):
    print("El punto está sobre el eje vertical")
else:
    print("El punto está en el cuadrante 4")
```

# Estructura match

Similar a la estructura switch-case de C++.
Ejemplo: Clasificar si un punto esta en el origen o en algun eje.

```python
coordenada = (0, 1) #Cambiar valores
match coordenada:
    case (0, 0):
        print("Coordenada en origen")
    case (x, 0):
        print(f"Coordenada en eje x: {x}")
    case (0, y):
        print(f"Coordenada en eje y: {y}")
    case (x, y):
        print(f"Coordenada en: {x}, {y}")
    case _:
        print("Error")
```
