El polimorfismo de herencia (method overriding) es la capacidad que tienen las **clases hijas** de proporcionar una implementación específica de un método que ya ha sido definido en su **clase padre**.

![Polimorfismo2](https://pynative.com/wp-content/uploads/2021/08/polymorphism_with_inheritance.jpg)

Debido a que Python es un lenguaje de tipado dinamico, no es necesario hacer el polimorfismo por herencia. En lenguajes de tipado estatico como Java, si es necesario.

```python
class Animal:
  def sonido(self):
    return 'No hace ruidos'

class Gato(Animal):
  def sonido(self):
    return "Miau"

class Perro(Animal):
  def sonido(self):
    return "Guau"

class Gusano(Animal):
  pass
```

## **Ejemplo:**

* Escribir el codigo de una **clase padre** `Vehiculo`, y reciba el nombre, color y precio.
* Agregar el **metodo** `mostrar` que imprime "Detalles: {nombre}, {color}, {precio}" del objeto.
* Agregar el **método** `cambios` que indique la cantidad de cambios que tiene el auto (5 por defecto).
* Agregar el **método estático** `velocidad` que indique la velocidad máxima del auto (150 km/h)

```python
class Vehiculo:
  def __init__(self, nombre, color, precio):
    self.nombre = nombre
    self.color = color
    self.precio = precio
  def mostrar(self):
    print(f"Detalles: {self.nombre}, {self.color}, {self.precio}")
  def cambios(self):
    print("El auto",self.nombre," tiene 5 cambios")
  @staticmethod
  def velocidad():
    print("Maximo 150 km/h")
```

```python
mi_carro = Vehiculo('Spark','Gris',12000000)
```

```python
mi_carro.mostrar()
mi_carro.cambios()
mi_carro.velocidad()
```

### **NOTA:** Métodos estáticos
* El decorador `@staticmethod` de indica que el método (en este caso `velocidad`) no depende de una instancia específica (un objeto individual) ni de la clase en sí para funcionar
* Python no envía nada. Es una función "ciega" a los datos del objeto. No sabe si el coche es azul, rojo o si costó un millón.

```python
class Camion(Vehiculo):
  def __init__(self, nombre, color, precio):
    super().__init__(nombre, color, precio)
  def velocidad(self):
    print("Maximo 200 km/h")
  def cambios(self):
    print("8 cambios")
```

```python
cam = Camion('Foton','Negro',120000000)
```

```python
cam.velocidad()
```

## **9.2. Encapsulacion**
*Consiste en agrupar un conjunto de datos y metodos en una sola unidad*. Entonces, por ejemplo, cuando se crea una clase, significa que estas implementando la encapsulacion. Una clase es un ejemplo de encapsulacion debido que enlaza todos los datos (variables de la instancia) y los metodos en una sola unidad.

![](https://pynative.com/wp-content/uploads/2021/08/encapsulation_python_class.jpg)

## **Ejemplo:**
En este ejemplo, creamos la clase `Employee` (empleado), definiendo los atributos del empleado como variables de instancia `name`, `salary` y `project`.
Además se implementaron los metodos `work`, `show` y `bonus` como metodos de la instancia.



```python
class Employee:
    def __init__(self, name, salary, project):
        self.name = name
        self.salary = salary
        self.project = project
    def show(self):
        print(" - Name: ", self.name, '\n - Salary:', self.salary, '\n - Bonus:', self.bonus())
    def work(self):
        print(self.name, 'is working on', self.project)
    def bonus(self):
        return self.salary*0.15
```

```python
emp1 = Employee('Jessa', 8000, 'NLP')
```

* Llamar los atributos públicos

```python
print(emp1.salary)
```

* Llamar los métodos públicos

```python
emp1.show()
emp1.work()
emp1.bonus()
```

## **9.3. Miembros protegidos y acceso de modificadores**
**Los metodos y atributos pueden ser públicos, privados o protegidos**. Para definirlo se usa el underscore (raya al piso) y doble underscore.

Los modificadores de acceso limitan el acceso a las variables y metodos de una clase. Python provee tres tipos de acceso a los modificadores: privados, publicos y protegidos.
* **Miembro Público:** Accesible desde cualquier parte por fuera de la clase.
* **Miembro Protegido:** Accesible dentro de la clase y sus subclases.
* **Miembro Privado:** Accesible dentro de la clase.


![](https://pynative.com/wp-content/uploads/2021/08/python_data_hiding.jpg)


## **9.3.1. Miembros privados**
Podemos proteger las variables en la clase volviendolas privadas. Usamos la doble raya al piso (`__`) como prefijo al comienzo del nombre de la variable.
Los miembros privados solo se pueden acceder dentro de la clase, pero no podemos acceder a ellos directamente desde los objetos de la clase.

* Hacer un atributo privado

```python
class Employee:
    def __init__(self, name, salary, project):
        self.name = name
        self.__salary = salary       # El atributo salary ahora es privado
        self.project = project
    def show(self):
        print(" - Name: ", self.name, '\n - Salary:', self.__salary, '\n - Bonus:', self.bonus())
    def work(self):
        print(self.name, 'is working on', self.project)
    def bonus(self):
        return self.__salary*0.15

emp2 = Employee('Jorge', 12000, 'G8')
```

* ¿Qué pasa si intentamomos acceder a un atributo privado?

```python
print(emp2.__salary)
```

* Podemos acceder al miembro privado usando un metodo de la instancia.

```python
emp2.show()
emp2.work()
emp2.bonus()
```

* Hacer un método privado

```python
class Employee:
    def __init__(self, name, salary, project):
        self.name = name
        self.__salary = salary
        self.project = project
    def show(self):
        print(" - Name: ", self.name, '\n - Salary:', self.__salary, '\n - Bonus:', self.__bonus())
    def work(self):
        print(self.name, 'is working on', self.project)
    def __bonus(self):               # El método bonus ahora es privado
        return self.__salary*0.15

emp3 = Employee('Erick', 9000, 'G8')
```

```python
emp3.__bonus()
```

```python
emp3.show()
```

## **9.3.2. Miembros protegidos**
Los miembros protegidos son accesibles dentro de la clase y tambien dentro de las sub-clases. Los miembros protegidos se definen iniciando con raya al piso (`_`).

```python
# Clase Padre
class Company:
    def __init__(self):
        self._project = "NLP"                    # Protected member
        self.__secret = "Nadie lo sabe"          # Private member

# Clase Hija
class Employee(Company):
    def __init__(self, name):
        self.name = name
        Company.__init__(self)
    def show(self):
        print("Employee name :", self.name)            # Accessing protected member in child class
        print("Working on project :", self._project)
        #print("Working on project :", self.__secret)

ITM = Company()
emp4 = Employee("Jessa")
```

```python
emp4.show()
```

```python
# Direct access protected data member
print('Project:', emp4._project)
```

```python
emp4._project = 'G8'

print(emp4._project)
```

```python
ITM._project = 'Proyecto'
ITM._project
```

## **9.4. Métodos Getters y Setters en Python**

Los Getters y Setters son métodos específicos que te permiten leer (`get`) o modificar (`set`) el valor de un atributo privado de forma controlada.

En lugar de dejar que cualquiera toque tus variables directamente, los obligas a pasar por estos métodos para asegurarte de que todo esté en orden.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.__age = age      # Atributo privado
    # Método Getter
    def get_age(self):
        return self.__age
    # Método Setter
    def set_age(self, age):
        self.__age = age

stud1 = Student('Jessa', 18)
```

```python
stud1.__age
```

* Llamar la edad usando getter

```python
print('Name:', stud1.name, 'has', stud1.get_age())
```

* Modificar la edad usando getter

```python
stud1.set_age(19)

print('Name:', stud1.name, 'has', stud1.get_age())
```

* Otra forma de hacerlo, usando el decorador `@property`

A los programadores de Python no les gusta escribir `objeto.get_valor()`. Muchos prefieren escribir `objeto.valor` como si fuera un atributo normal, pero que por detrás ejecute un método. Para eso usamos decoradores.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Atributo privado
    # Método Getter
    @property
    def age(self):
        return self.__age
    # Método Setter
    @age.setter
    def age(self, value):
        self.__age = value

stud2 = Student('Juan', 21)
```

```python
print('Name:', stud2.name, 'has', stud2.age)
```

```python
stud2.age = 19

print('Name:', stud2.name, 'has', stud2.age)
```

## **Ventajas de la Encapsulacion**
* **La seguridad:** La principal ventaja de utilizar la encapsulación es la seguridad de los datos. La encapsulación protege un objeto de accesos no autorizados. Permite niveles de acceso privados y protegidos para evitar la modificación accidental de los datos.

* **Ocultación de datos:** El usuario no sabría lo que está pasando detrás de la escena. Sólo sabría que para modificar un miembro de datos, hay que llamar al método setter. Para leer un miembro de datos, llame al método getter. Lo que estos métodos setter y getter están haciendo está oculto para ellos.

* **Simplicidad:** Simplifica el mantenimiento de la aplicación al mantener las clases separadas y evitar que se acoplen estrechamente entre sí.

* **Estética:** Agrupar datos y métodos dentro de una clase hace que el código sea más legible y fácil de mantener.

## **9.5.2. Metodos de tipo clase**
Se definen a partir del decorador `@classmethod`. A diferencia de los métodos de instancia, estos métodos de clase reciben como argumento _cls_, que hace referencia a la clase. Por lo tanto, pueden acceder a la clase pero no a la instancia. Dicho de otra forma, no pueden acceder a los atributos de la instancia ya que no reciben `self`, pero si pueden recibir y posteriormente modificar los atributos de la clase.

```python
class Usuario:
  def __init__(self, nombre='Pepito'):
    self.nombre = nombre
    self.edad = 100
    self.sexo = 'no binario'
    self.birthday = Fecha()
  def __str__(self):
    return ' nombre: '+self.nombre+'\n Edad: '+str(self.edad)+'\n Sexo: '+str(self.sexo)+\
        '\n Fecha de nacimiento: '+str(self.birthday)
  def EdadInstancia(self):
    return self.edad
  @classmethod
  def EdadClase(cls):
    return cls().edad
```

```python
JuanSe = Usuario()
JuanSe.edad = 21
JuanSeClonEdad = JuanSe.EdadInstancia()
UsuarioEdad = JuanSe.EdadClase()
#Pepito = JuanSeClase()
```

```python
print(JuanSe.EdadInstancia(), JuanSe.EdadClase())
print(JuanSeClonEdad,UsuarioEdad)
```

## **9.5.3. Metodos estaticos**
Se definen a partir del decorador @staticmethod. No reciben como argumento de entrada ni la clase ni la instancia. Por tal motivo no pueden modificar el estado de la clase ni de la instancia.

```python
class Usuario:
  def __init__(self, nombre='Pepito'):
    self.nombre = nombre
    self.edad = 100
    self.sexo = 'no binario'
    self.birthday = Fecha()
  def __str__(self):
    return ' nombre: '+self.nombre+'\n Edad: '+str(self.edad)+'\n Sexo: '+str(self.sexo)+\
        '\n Fecha de nacimiento: '+str(self.birthday)
  @staticmethod
  def metodoestatico(x,y):
    return "Soy independiente de la instancia y de la clase, soy un "+str(x+y)
```

```python
Samuel = Usuario()
print(Samuel.metodoestatico(5,5))
```
