### **10.2.1. Arreglos (arrays)**

Un arreglo (array) es una estructura de datos diseñada para almacenar una colección de elementos (generalmente números) del mismo tipo en posiciones de memoria contiguas.

### **Características fundamentales**

* **Homogeneidad**

A diferencia de una lista de Python (que puede mezclar peras con manzanas), un arreglo exige que todos sus datos sean iguales (por ejemplo, que todos sean float64). Esto permite que la computadora procese los datos mucho más rápido.

* **Eficiencia de memoria**

Al estar "pegados" unos con otros en la RAM, el procesador puede acceder a ellos de forma secuencial sin saltos innecesarios, lo que optimiza el uso de la caché de tu CPU.

* **Dimensiones**

1D: Un vector (una fila de datos).

2D: Una matriz (filas y columnas).

3D o más: Tensores (útiles para datos volumétricos o series temporales).

* Técnicamente, se acaba de transformar un vector (1D) en una matriz columna (2D).

```python
L1 = [1,2,3,4]
L2 = [5,3,2,19]
print(type(L1), type(L2))

#Sumar estas dos listas
R = L1 + L2
print("Resultado:",R)
```

```python
A1 = np.array(L1)
A2 = np.array(L2)
print(A1)
print(A2)

#Sumar estos dos arreglos
Rn = A1 + A2
print("Resultado:",Rn)
```

* Podemos crear una función que sume las listas como su fueran arreglos:

```python
def SumaLista(L1,L2):
    R = []
    for i in range(len(L1)):
        R.append(L1[i]+L2[i])

    return R
```

```python
SumaLista(L1,L2)
```

```python
SumaLista([3,5,2,1,2,4,6,2],[1,2,4,56,8,9,1,2])
```

```python
Lista1 = list(range(1000000))
Lista2 = list(range(1000000))
```

Sin embargo, la función `SumaLista` es mucho menos eficiente que la suma de arreglos.

Para demostrarlo usaremos `%timeit`, que se utiliza para medir el tiempo de ejecución de una línea de código de forma precisa y estadística.

```python
%%timeit
RL = SumaLista(Lista1,Lista2)
```

```python
Arr1 = np.array(Lista1)
Arr2 = np.array(Lista2)
```

```python
%%timeit
RA = Arr1 + Arr2
```

```python
RL = SumaLista(Lista1,Lista2)
RA = Arr1 + Arr2
print(RA)
```

```python
RA == RL
```

```python
np.all(RA == RL)
```

- $k$: Índice de la diagonal. 0 es la principal (por defecto), valores positivos suben la diagonal y negativos la bajan.

## **Ejemplo:**

La definición formal de un número aleatorio en un rango es: $(b-a)*rand + a$. Utilicemos esta definición para crear una matriz aleatoria cuyos elementos tengan valores entre 5 y 15.

```python
import random
#seed o semilla
M = np.zeros([3,5])

a = 5
b = 15

for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i,j] = (b-a)*random.random()+a       # random.random() genera un único valor aleatorio
print(M)
```

```python
M2 = (b-a)*np.random.rand(4,4)+a
print(M)
```

* La función sigmoide es una función matemática con forma de “S” que se usa mucho en áreas como el aprendizaje automático, la ciencia de datos, estadística y biología.
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$
Esta función no está incluida en los módulos math o numpy. Por tanto, vamos a crear la función sigmoide:

```python
def sigmoid(z):
    return 1. / (1. + math.exp(-z))
```

```python
x = np.linspace(-10,10,21)
print(x)
print(x.shape)
```

```python
R = []
for i in x:
    R.append(sigmoid(i))
```

```python
plt.plot(x,R)
plt.grid()
#plt.xlim([-5,5])
```

### **10.3.3. Broadcasting**
El **broadcasting** es una de las características más potentes de NumPy. Es lo que permite realizar operaciones aritméticas entre arrays que tienen diferentes formas (shapes).

Normalmente, para sumar dos matrices, tendrían que ser del mismo tamaño. El **broadcasting** rompe esa regla siguiendo un conjunto de normas lógicas para "estirar" el array más pequeño de modo que coincida con el más grande.

```python
np.random.seed(2)

A0 = np.random.randint(0,4,(4,3))
print(A0)
print(f"Tamaño de A0= {A0.shape}")
```

```python
A1 = np.ones((4,1), dtype= int)

print(A1)
print(f"Tamaño de A1= {A1.shape}")
```

```python
print(A0 + A1)
```

Para saber si dos arrays son compatibles, NumPy compara sus dimensiones de derecha a izquierda (empezando por la última). Dos dimensiones son compatibles si:

 - Son iguales.

- Una de ellas es 1.

```python
np.random.seed(2)

Ak = np.random.randint(0,4,(5,4))
print(Ak)
print(f"\nTamaño de Ak = {Ak.shape}\n")
media_fila = np.mean(Ak, axis=1)
print(media_fila)
print(f"\nTamaño de media_fila = {media_fila.shape}")
```

```python
Akm = Ak - np.mean(Ak, axis=1)
print(np.mean(Ak, axis=1))
print(Akm)
```
