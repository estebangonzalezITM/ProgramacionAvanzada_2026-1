### **10.2.1. Arreglos (arrays)**

Un arreglo (array) es una estructura de datos diseñada para almacenar una colección de elementos (generalmente números) del mismo tipo en posiciones de memoria contiguas.

### **Características fundamentales**

* **Homogeneidad**

A diferencia de una lista de Python (que puede mezclar peras con manzanas), un arreglo exige que todos sus datos sean iguales (por ejemplo, que todos sean float64). Esto permite que la computadora procese los datos mucho más rápido.

* **Eficiencia de memoria**

Al estar "pegados" unos con otros en la RAM, el procesador puede acceder a ellos de forma secuencial sin saltos innecesarios, lo que optimiza el uso de la caché de tu CPU.

* **Dimensiones**

1D: Un vector (una fila de datos).

2D: Una matriz (filas y columnas).

3D o más: Tensores (útiles para datos volumétricos o series temporales).

* Técnicamente, se acaba de transformar un vector (1D) en una matriz columna (2D).

```python
L1 = [1,2,3,4]
L2 = [5,3,2,19]
print(type(L1), type(L2))

#Sumar estas dos listas
R = L1 + L2
print("Resultado:",R)
```

```python
L1n = np.array(L1)
L2n = np.array(L2)
print(L1n)
print(L2n)

#Sumar estos dos arreglos
Rn = L1n + L2n
print("Resultado:",Rn)
```

* Podemos crear una función que sume las listas como su fueran arreglos:

```python
def SumaLista(L1,L2):
    R = []
    for i in range(len(L1)):
        R.append(L1[i]+L2[i])

    return R
```

```python
SumaLista(L1,L2)
```

```python
SumaLista([3,5,2,1,2,4,6,2],[1,2,4,56,8,9,1,2])
```

- $M$: (Opcional) Número de columnas. Si no se pone, la matriz será cuadrada ($N \times N$).
- $k$: Índice de la diagonal. 0 es la principal (por defecto), valores positivos suben la diagonal y negativos la bajan.

La definición formal de un número aleatorio en un rango es: $(b-a)*rand + a$. Utilicemos esta definición para crear una matriz aleatoria cuyos elementos tengan valores entre 5 y 15.

```python
import random
#seed o semilla
M = np.zeros([5,5])

a = 5
b = 15

for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i,j] = (b-a)*random.random()+a
print(M)
```

Numpy contiene las versiones vectorizadas de las funciones de la librería math, por ejemplo:
- math.sin --> numpy.sin
- math.exp --> numpy.exp

A diferencia de las funciones de la librería math, las versiones vectorizadas pueden recibir arreglos n-dimensionales como argumentos.

```python
import math

def sigmoid(z):
    return 1. / (1. + math.exp(-z))
```
