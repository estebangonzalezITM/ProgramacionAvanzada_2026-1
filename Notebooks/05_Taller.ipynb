```python
#Este código identifica si un numéro es divisible por 5 o por 11.
Numero = int(input("Ingrese un número: "))
if Numero % 5 == 0:
    print("El número es divisible por 5.")
elif Numero % 11 == 0:
    print("El número es divisible por 11.")
else:
    print("El número no es divisible ni por 5 ni por 11.")
```

```python
#Este programa identifica si un caracter es alfabetico, numérico o un espacio.
Caracter = input("Ingrese un caracter: ")
if Caracter.isalpha():
    print("El caracter es alfabetico.")
elif Caracter.isdigit():
    print("El caracter es numérico.")
elif Caracter.isspace():
    print("El caracter es un espacio.")
else:
    print("El caracter no es ni alfabetico, ni numérico, ni un espacio.")
```

```python
#Este programa solicita si un caracter es una vocal o una consonante.
Caracter = input("Ingrese un caracter: ")
if Caracter.lower() in ['a', 'e', 'i', 'o', 'u']:
    print("El caracter es una vocal.")
elif Caracter.isalpha():
    print("El caracter es una consonante.")
else:
    print("El caracter no es una letra.")   
```

```python
#Programa que solicita un mes y devuelve el número de días que tiene ese mes.
Mes = int(input("Ingrese el número de un mes o ingrese 0 para salir: "))
while Mes != 0:
    if Mes in [1, 3, 5, 7, 8, 10, 12]:
        print(f"El mes {Mes} tiene 31 días.")
    elif Mes in [4, 6, 9, 11]:
        print(f"El mes {Mes} tiene 30 días.")
    elif Mes == 2:
        print(f"El mes {Mes} tiene 28 o 29 días, dependiendo si es un año bisiesto.")
    elif Mes >12:
        print("El número del mes no es válido.")
    Mes = int(input("Ingrese el número de un mes o ingrese 0 para salir: "))
else:
    print("Gracias por usar el programa.")
```

```python
#programa que verifica si la suma de los angulos de un triangulo es igual a 180 grados.
Angulo1 = float(input("Ingrese el primer ángulo del triángulo: "))
Angulo2 = float(input("Ingrese el segundo ángulo del triángulo: "))
Angulo3 = float(input("Ingrese el tercer ángulo del triángulo: "))
SumaAngulos = Angulo1 + Angulo2 + Angulo3
if SumaAngulos == 180:
    print("La suma de los ángulos es igual a 180 grados. Es un triángulo válido.")
else:
    print("La suma de los ángulos no es igual a 180 grados. No es un triángulo válido.")    
```

```python
#programa que solicita los lados de un triangulo y verifica si es valido o no.
lado1 = float(input("Ingrese el primer lado del triángulo: "))
lado2 = float(input("Ingrese el segundo lado del triángulo: "))
lado3 = float(input("Ingrese el tercer lado del triángulo: "))
if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Los lados ingresados pueden formar un triángulo válido.")
else:
    print("Los lados ingresados NO pueden formar un triángulo válido.")
```
