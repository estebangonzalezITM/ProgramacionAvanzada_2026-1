## **8.2. Clases (El molde)**

La programación orientada a objetos en Python organiza el código mediante class (clases) para crear modelos de objetos que agrupan atributos (datos) y métodos (comportamientos).

Una clase es una plantilla o un "plano" que define la estructura y el comportamiento de algo. No es el objeto real, sino las instrucciones de cómo construirlo.

Por ejemplo, el número `a = 3` es de clase 'int', mientras que el número `b = 3.0` es de clase 'float'.


```python
a = 3  ; b= 3.0
print('La clase de a es: ',type(a))
print('La clase de b es: ',type(b))
```

* La clase determina cuáles serán los métodos y atributos de un objeto.
* En este ejemplo, ´a´ y ´b´ NO son clases, son instancias u objetos

## **8.4. Atributos y Métodos**
### **8.4.1 Atributos**
Todas variables o atributos de la clase, se guardan dentro de la clase por medio del objeto `self` (en español "uno mismo"). Tambien se pueden definir los atributos inmediatamente despues de definir el nombre de la clase y su tipo.

```python
class Clase1:
  x = 2
  def __init__(self):
    self.y = self.x + 1
```



* `x= 2` es un atributo de la Clase1. Es una variable compartida por todos los objetos que se creen de esa clase
*  `__init__` es el método constructor, que actúa como la puerta de entrada para configurar el objeto.
Sin él, el objeto no sabría qué valores iniciales tomar al nacer.
* `self` es una referencia al objeto específico que se está creando, permitiéndole acceder a sus propios datos
* `self.x` en un atributo de clase. Esta línea busca el valor de `x` (que es 2) para realizar el cálculo inicial del objeto.
* `self.y` es un atributo de instancia, es una variable única para cada objeto individual.




```python
instancia1 = Clase1()
print(instancia1.x, instancia1.y)
```

* `instancia1` es un objeto (instancia) creada a partir del **molde** Clase1.
* `instancia1.x` es un **Atributo de Clase**. Pertenece a la clase y todas las instancias lo comparten (inicialmente).
* `instancia1.y` es un  **Atributo de Instancia**. Pertenece exclusivamente a `instancia1` y se crea en el momento en que el objeto "nace".

```python
print(type(Clase1))
print(type(instancia1))
```

* Analicemos el siguiente ejemplo:

```python
instancia1.x = 4
print(instancia1.x,instancia1.y)
```

Modificar un atribujo no 'inicia' el método constructor, por tanto, `instancia.y` no cambia.

### **8.4.4. Metodos**
Los metodos son las funciones que se definen dentro de la clase. Los metodos pueden emplear atributos o argumentos externos al momento de ser invocados. Por ejemplo creamos una clase llamada `vector2d` que almacena las componentes en los ejes horizontal y vertical. Posteriormente creamos un medodo (una funcional dentro de la clase) que reciba un vector y los sume con la instancia.

```python
class vector2d:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def suma(self, vector):
    # Sumar dos Vectores
    return vector2d(self.x + vector.x, self.y + vector.y)
```

```python
vector1 = vector2d(3,4)
vector2 = vector2d(4,3)
vector3 = vector1.suma(vector2)
print(f"La suman en el eje x es: {vector3.x}, y en el eje y es: {vector3.y}")
```

## 4. Sobrecarga de operadores
Podemos definir que operacion realizar de acuerdo a la operacion que se este realizando con una instancia de la clase. Por ejemplo definir que pasa con b + c, si las variables b y c son de una clase nueva. <br>


```python
class vector2d:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def __add__ (self, vector):
    # Sumar dos Vectores
    return vector2d(self.x + vector.x, self.y + vector.y)
```

```python
a = vector2d(2,3)
b = vector2d(4,5)
c = a+b
print(c.x,', ',c.y)
```

## 5. Herencia
Es un concepto fundamental en la programacion orientada a objetos. La idea es crear clases u objetos a partir de otra clase, y heredar los atributos y metodos de la clase madre/padre. Como ejemplo creamos una clase animal y una subclase Perro:

```python
# Definimos una clase padre
class Animal:
    pass

# Creamos una clase hija que hereda de la padre
class Perro(Animal):
    pass
```

Podemos ver sí la clase `Perro` es subclase de la clase `Animal`:

```python
print(Perro.__bases__)
```

Y podemos imprimir las subclases asociadas a la clase `Animal`

```python
print(Animal.__subclasses__())
```

### 5.1 Extensión y modificación de métodos de una clase padre

Creamos la clase general para los animales, con atributos `specie` y `age` para el tipo de especie y la edad del animal. Adicionalmente, los metodos `speak`, `move` y `describe`, como se define a continuación:

```python
class Animal:
    def __init__(self, specie, age):
        self.specie = specie
        self.age = age

    # Método genérico pero con implementación particular
    def speak(self):
        # Método vacío
        pass

    # Método genérico pero con implementación particular
    def move(self):
        # Método vacío
        pass

    # Método genérico con la misma implementación
    def describe(self):
        print("Soy un Animal del tipo", type(self).__name__)
```

Ahora creamos dos clases hijas a partir del padre `Animal` y modificamos los metodos:

```python
class Perro(Animal):
    def hablar(self):
        print("Guau!")
    def moverse(self):
        print("Caminando con 4 patas")

class Pollo(Animal):
    def hablar(self):
        print("Pio! Pio!")
    def moverse(self):
        print("Caminando con 2 patas")
```

### 5.2 Uso de `super()`

El uso de `super()` nos permite acceder a los metodos de la clase padre desde una clase hija.


```python
class Animal:
    def __init__(self, especie, edad):
        self.especie = especie
        self.edad = edad
    def hablar(self):
        pass

    def moverse(self):
        pass

    def describeme(self):
        print("Soy un Animal del tipo", type(self).__name__)
```

Tal vez queramos que nuestro Perro tenga un parámetro extra en el constructor, como podría ser el dueño. Para realizar esto tenemos dos alternativas:
- Podemos crear un nuevo `init` y guardar todas las variables una a una.
- O podemos usar super() para llamar al `init` de la clase padre que ya aceptaba la especie y edad, y sólo asignar la variable nueva manualmente.

```python
class Perro(Animal):
  def __init__(self, especie, edad, dueño=Usuario()):
    # Alternativa 1
    self.especie = especie
    self.edad = edad
    self.dueño = dueño

class Perro2(Animal):
  def __init__(self, especie, edad, dueño):
    # Alternativa 2
    super().__init__(especie, edad)
    self.dueño = dueño
```

```python
streeter = Perro('mamifero',5)
```

```python
print(streeter.dueño)
```

### 5.3 Herencia multiple

Tambien se puede realizar herencia a partir de mas de una clase. Por ejemplo,  creamos la clase Padre y Madre, y construimos una clase Hijo que herede los metodos y atributos de Padre y Madre.

```python
# Definicion clase para Padre
class Padre():
	# Metodo de la clase Padre
	def canta(self):
		print("Papá canta bien.")

# Definicion clase para Madre
class Madre():
	# Metodo de la clase Madre
	def programa(self):
		print("Mamá programa bien.")

# Definicion de la clase derivada
class Hijo(Padre, Madre):
	def jugar(self):
		print("Al niño le encanta jugar")

# creando un objeto de la clase derivada
niño = Hijo()
# llamando a los metodos de las clases padres y la derivada
niño.canta()
niño.programa()
niño.jugar()
```
